```python
import pysrt

# Load subtitles from an .srt file
subtitles = pysrt.open('/content/drive/My Drive/hack/Just go with it subtitle.srt', encoding='utf-8')
text_lines = [sub.text for sub in subtitles]
```

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification
import torch

# Load pre-trained model and tokenizer
model_name = "j-hartmann/emotion-english-roberta-large"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = RobertaForSequenceClassification.from_pretrained(model_name)

# Analyze sentiments for each line
subtitle_sentiments = []
for line in text_lines:
    inputs = tokenizer(line, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()
    subtitle_sentiments.append((line, predicted_class))
```

```python
# Filter subtitles for "Joy" sentiment (comedy scenes) and keep track of timestamps
comedy_segments = []
for subtitle, sentiment in subtitle_sentiments:
    # Assuming 2 represents 'Joy' sentiment
    if sentiment == 6 :
        comedy_segments.append(subtitle)
```

```python
print (comedy_segments)
```

```python
# Collect start and end times of each comedy scene
timestamps = [(sub.start.to_time(), sub.end.to_time()) for sub, sentiment in zip(subtitles, subtitle_sentiments) if sentiment == 2]
```

```python
import subprocess

# Assume 'input_movie.mp4' is the movie file and 'timestamps' has start and end times
input_movie = 'input_movie.mp4'
for i, (start, end) in enumerate(timestamps):
    # Convert start and end to HH:MM:SS format if necessary
    start_time = start.strftime('%H:%M:%S')
    end_time = end.strftime('%H:%M:%S')

    # Output file name for each segment
    output_file = f'comedy_scene_{i}.mp4'

    # FFmpeg command to extract segment
    command = [
        'ffmpeg', '-i', input_movie, '-ss', start_time, '-to', end_time,
        '-c', 'copy', output_file
    ]
    subprocess.run(command)
```
